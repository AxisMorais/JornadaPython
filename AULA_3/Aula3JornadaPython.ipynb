```python
import pandas as pd 
tabela = pd.read_csv("C:/Users/coneg/OneDrive/Área de Trabalho/JornadaPython/AULA_3/clientes.csv")
display(tabela)
```

```python
print(tabela.info())
print(tabela.columns)
```

```python
from sklearn.preprocessing import LabelEncoder
#Transforma a coluna de texto em número 
codificador = LabelEncoder()

#Aplicamos o codificador em todas as colunas exceto na coluna score_credito
for coluna in tabela.columns:
    if tabela[coluna].dtype == "object" and coluna != "score_credito":
        tabela[coluna] = codificador.fit_transform(tabela[coluna])
```

```python
tabela["profissao"] = codificador.fit_transform(tabela["profissao"])
```

```python
print(tabela.info())
```

```python
# A INTELIGÊNCIA ARTIFICIAL PRECISA DE DADOS DE TREINO E DADOS DE TESTE

y = tabela["score_credito"]
x = tabela.drop(columns =["score_credito", "id_cliente"])

from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y)
```

```python
print(x_treino)
```

```python
#Criar a Inteligência artificial
#Criar o modelo de Arvore de Decisão
#KNN - Vizinhos próximos -> Kneighbors
```

```python
#Importa a inteligênfcia artificial 
#Cria a inteligência e treina

#Importa 
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

#Cria
modelo_arvoredecisao = RandomForestClassifier()
modelo_knn = KNeighborsClassifier()

#Treina
modelo_arvoredecisao.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino)
```

```python

from sklearn.metrics import accuracy_score

previsao_arvoredecisao = modelo_arvoredecisao.predict(x_teste) 
previsao_knn = modelo_knn.predict(x_teste.to_numpy())

print(accuracy_score(y_teste, previsao_arvoredecisao))
print(accuracy_score(y_teste, previsao_knn))
```
